```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=bfabbb93719f6e3e0937f22cb6505c70&language=en-US&page=1)')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
temp_df.head()
```

```python
data_frames = []

for i in range(1, 500):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=bfabbb93719f6e3e0937f22cb6505c70&language=en-US&page={}'.format(i))
    response_data = response.json()

    if 'results' in response_data:
        temp_df = pd.DataFrame(response_data['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        data_frames.append(temp_df)
    else:
        print(f"No results found in response for page {i}")

df = pd.concat(data_frames, ignore_index=True)
```
